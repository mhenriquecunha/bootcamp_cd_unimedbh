# Calcule a média de preço por marca e por veículo

```python
# Média de preço por marca e por veículo
trial = pd.DataFrame()
for b in list(df["marca"].unique()):
    for v in list(df["tp_de_veiculo"].unique()):
        z = df[(df["marca"] == b) & (df["tp_de_veiculo"] == v)]["preço"].mean()
        trial = trial.append(pd.DataFrame({'marca':b , 'tp_de_veiculo':v , 'avgpreço':z}, index=[0]))
trial = trial.reset_index()
del trial["index"]
trial["avgpreço"].fillna(0,inplace=True)
trial["avgpreço"].isnull().value_counts()
trial["avgpreço"] = trial["avgpreço"].astype(int)
trial.head(5)
```

# Preço médio de um veículo por marca, bem como tipo de veículo

```python
# Crie um Mapa de Calor com Preço médio de um veículo por marca, bem como tipo de veículo
tri = trial.pivot("marca","tp_de_veiculo", "avgpreço")
fig, ax = plt.subplots(figsize=(15,20))
sns.heatmap(tri,linewidths=1,cmap="YlGnBu",annot=True, ax=ax, fmt="d")
ax.set_title("Mapa de Calor - Preço médio de um veículo por marca e tipo de veículo",fontdict={'size':20})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':20})
ax.yaxis.set_label_text("Marca",fontdict= {'size':20})
plt.show()
```
